```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from bisect import bisect_right
```

```python
file_path = "./Datasets/all_accelerometer_data_pids_13.csv"
df = pd.read_csv(file_path)
```

```python
df.describe
```

```python
pids = pd.unique(df['pid'])
pids
acc={}

for pid in pids:
  acc[pid] = df.loc[df['pid']==pid]
  acc[pid] = acc[pid][acc[pid]['time']!=0]
  acc[pid]['time'] //= 1000
  acc[pid]['datetime'] = pd.to_datetime(acc[pid]['time'], unit='s')
  acc[pid].sort_values(by=['time'])
```

```python
merged={}
tac={}
for pid in pids:
  tac_path ="./Datasets/Preprocessed/clean_tac/"+ pid +'_clean_TAC.csv'
  tac[pid] = pd.read_csv(tac_path)
  tac[pid].sort_values(by=['timestamp'])
  merged[pid] = pd.merge(acc[pid], tac[pid], how='inner', left_on='time', right_on='timestamp')
  tac[pid]['datetime'] = pd.to_datetime(tac[pid]['timestamp'], unit='s')
```

```python
 plt.figure(figsize=(20,5))
    # Create datetime for reference in plot.

for pid in pids:
    plt.plot(tac[pid]["datetime"], tac[pid]["TAC_Reading"], marker="o", linestyle="", label=pid)
plt.axhline(y=0.08, color='r', linestyle='-')
plt.title("TAC Readings Over Time By Subject")
plt.xlabel("Datetime")
plt.legend()
plt.show()
```
